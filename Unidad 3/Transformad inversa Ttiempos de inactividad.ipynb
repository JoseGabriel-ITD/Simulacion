### Método de la transformada inversa 
    
- El método de la transformada inversa se utiliza para generar variables aleatorias que siguen una distribución específica, **transformando números aleatorios uniformes 𝑈 (0,1) en variables con la distribución deseada.**


### Distribución exponencial 

- La distribución exponencial **es una distribución de probabilidad continua que se utiliza para modelar el tiempo entre eventos, donde los eventos ocurren de manera independiente a una tasa constante.**
- Describe la probabilidad de que pase un cierto período de tiempo antes de que ocurra un evento. Algunas características clave de la distribución exponencial incluyen:

1. **Parámetro de tasa ($\lambda$)**: La distribución exponencial está completamente definida por un parámetro de tasa ($\lambda$), que representa la tasa promedio de eventos por unidad de tiempo. Cuanto mayor sea ($λ$), más eventos ocurrirán en un período de tiempo dado.

2. **Función de densidad de probabilidad (PDF)**: La función de densidad de probabilidad de la distribución exponencial se define como:


$$
f(x) = \begin{cases} \lambda \cdot e^{-(λ\cdot x)}
\text{ para x >= 0} \\ 0
\text{ para x < 0} \end{cases} \therefore
$$

$$
\lambda \text { es el valor de la taza media}
$$

e = 2,7182818284590452…
$$









### Ejemplo de Aplicación:

#### Aplicar el método de la transformada inversa para generar variables aleatorias que simulen los tiempos de inactividad de un brazo de robot de soldadura en una fabrica de vehiculos


- Supongamos que el **tiempo promedio de inactividad del brazo de robot es de 10 minutos**, lo que implica que la **tasa de inactividad ($\lambda$) =  $\frac{1}{10}$ ya que ($\lambda$) es el inverso del tiempo promedio)**.

#### Contexto:
- Para simular los tiempos de inactividad de un brazo de robot de soldadura en una fábrica de vehículos, podemos **usar el método de la transformada inversa si asumimos que estos tiempos siguen una distribución conocida**. 
- En este ejemplo, supondremos que el tiempo entre inicios de inactividad (por ejemplo, debido a mantenimiento o a fallos) del brazo de robot sigue una distribución exponencial. 
- Un modelo común **para los tiempos de inactividad (o tiempos entre fallos) en sistemas industriales, como un brazo de robot de soldadura, es la distribución exponencial**. 
- Esto es razonable si se supone que los eventos que causan inactividad (como fallos o mantenimiento programado) ocurren de forma aleatoria e independiente con una tasa constante en el tiempo. 
- Esta distribución es adecuada para modelar los tiempos entre eventos en los que la probabilidad de ocurrencia es constante en el tiempo, como en el caso de fallos o inicios de inactividad.

   
    
*El método de la transformada inversa es utilizada para generar variables aleatorias que sigan distribuciones específicas, como la distribución exponencial utilizada en este caso para modelar los tiempos de inactividad de un brazo de robot de soldadura. Permite simular fenómenos de mantenimiento y fallos aleatorios en sistemas industriales, lo que es fundamental para la planificación de la producción y la gestión de recursos en fábricas*.

**Pasos para generar tiempos de inactividad usando el método de la transformada inversa:**
1. Generar un número aleatorio uniforme $u$ en el intervalo [0, 1].
2. Aplicar la fórmula para obtener 

$x$ =  -$\frac{1}{\lambda}$ln(1-$u$), donde ($\lambda$)  = $\frac{1}{10}$  
  
3. El valor de 𝑥 será el **tiempo de inactividad** simulado

**Supongamos que generamos el numero aleatorio $u$ = 0.65**

1. *Aplicamos la fórmula con:* $\lambda$ = $\frac{1}{10}$ 

$x$ =  - $\frac{1}{ \frac{1}{10}}$ ln (1 - 0.65) = -10 ln (0.35)  

  
2. *Calculamos:*  

$x$ $\approx$ -10 (-1.0498) $\approx$ 10.498 minutos

### Ejemplo aplicado en Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de inactividad)
lambda_inactividad = 1 / 10  # Tiempo promedio de inactividad = 10 minutos

# Función para generar un tiempo de inactividad
def generar_tiempo_inactividad(lambda_inactividad):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    tiempo_inactividad = -1 / lambda_inactividad * np.log(1 - u)  # Aplicar transformada inversa
    return tiempo_inactividad

# Generar 100 tiempos de inactividad
tiempos_inactividad = [generar_tiempo_inactividad(lambda_inactividad) for _ in range(100)]

# Mostrar los tiempos de inactividad generados
print("Tiempos de inactividad generados:", tiempos_inactividad)

# Graficar los tiempos de inactividad simulados
plt.hist(tiempos_inactividad, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos de Inactividad Simulados (Exponencial)')
plt.xlabel('Tiempo de inactividad (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
