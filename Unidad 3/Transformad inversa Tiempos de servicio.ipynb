### Método de la transformada inversa 
    
- El método de la transformada inversa se utiliza para generar variables aleatorias que siguen una distribución específica, **transformando números aleatorios uniformes 𝑈 (0,1) en variables con la distribución deseada.**


### Distribución exponencial 

- La distribución exponencial **es una distribución de probabilidad continua que se utiliza para modelar el tiempo entre eventos, donde los eventos ocurren de manera independiente a una tasa constante.**
- Describe la probabilidad de que pase un cierto período de tiempo antes de que ocurra un evento. Algunas características clave de la distribución exponencial incluyen:

1. **Parámetro de tasa ($\lambda$)**: La distribución exponencial está completamente definida por un parámetro de tasa ($\lambda$), que representa la tasa promedio de eventos por unidad de tiempo. Cuanto mayor sea ($λ$), más eventos ocurrirán en un período de tiempo dado.

2. **Función de densidad de probabilidad (PDF)**: La función de densidad de probabilidad de la distribución exponencial se define como:


$$
f(x) = \begin{cases} \lambda \cdot e^{-(λ\cdot x)}
\text{ para x >= 0} \\ 0
\text{ para x < 0} \end{cases} \therefore
$$

$$
\lambda \text { es el valor de la taza media}
$$

### Ejemplo de Aplicación:

#### Aplicar el método de la transformada inversa para generar variables aleatorias que simulen los tiempos de servicio en un restaurante de comida rápida

#### Contexto:

Para simular los tiempos de servicio en un restaurante de comida rápida, podemos suponer que estos tiempos siguen una distribución exponencial, que es común para modelar el tiempo entre eventos en sistemas de servicio.

**Tiempo de servicio** = una distribución exponencial con un parámetro **λ (la tasa de servicio)**. 

**La distribución exponencial** es adecuada porque describe situaciones en las que los tiempos entre eventos (en este caso, los tiempos de servicio) son independientes y ocurren de manera continua y aleatoria.


#### Proceso:
- Generación de números aleatorios: La función np.random.uniform(0, 1) genera un número aleatorio uniforme entre 0 y 1.
- Transformación: Usamos la fórmula del método de la transformada inversa para obtener los tiempos de servicio exponenciales.
- Visualización: Se genera un histograma que muestra la distribución de los tiempos de servicio simulados, lo cual sigue la forma esperada de la distribución exponencial.

**Pasos para generar tiempos de inactividad usando el método de la transformada inversa:**

Supongamos que **el tiempo promedio de servicio en el restaurante es de 2 minutos**, lo que implica que $\lambda$ = $\frac{1}{2}$

1. Generar un número aleatorio uniforme $u$ en el intervalo [0, 1].
2. Aplicar la fórmula para obtener 

$x$ =  -$\frac{1}{\lambda}$ln(1-$u$), donde ($\lambda$)  = $\frac{1}{2}$  
  
3. El valor de 𝑥 será el **tiempo de servicio** simulado  

**Supongamos que generamos el numero aleatorio $u$ = 0.73**

1. *Aplicamos la fórmula con:* $\lambda$ = $\frac{1}{2}$ 

$x$ =  - $\frac{1}{ \frac{1}{2}}$ ln (1 - 0.73) = -2 ln (0.27)  

2. *Calculamos:*  

$x$ $\approx$ -2 (0.27 $\approx$ 10.498 minutos

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de servicio)
lambda_servicio = 1 / 2  # Tiempo promedio de servicio = 2 minutos

# Función para generar un tiempo de servicio
def generar_tiempo_servicio(lambda_servicio):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    tiempo_servicio = -1 / lambda_servicio * np.log(1 - u)  # Aplicar transformada inversa
    return tiempo_servicio

# Generar 100 tiempos de servicio
tiempos_servicio = [generar_tiempo_servicio(lambda_servicio) for _ in range(100)]

# Mostrar los tiempos de servicio
print("Tiempos de servicio generados:", tiempos_servicio)

# Graficar los tiempos de servicio simulados
plt.hist(tiempos_servicio, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos de Servicio Simulados (Exponencial)')
plt.xlabel('Tiempo de servicio (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
