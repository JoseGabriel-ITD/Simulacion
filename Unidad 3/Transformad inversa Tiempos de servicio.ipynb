## Practica: Simular los tiempos de servicio en un restaurante de comida r谩pida

#### El m茅todo de la transformada inversa se utiliza para generar variables aleatorias que siguen una distribuci贸n espec铆fica, transformando n煤meros aleatorios uniformes  (0,1) en variables con la distribuci贸n deseada

#### Contexto:

Para simular los tiempos de servicio en un restaurante de comida r谩pida, podemos suponer que estos tiempos siguen una distribuci贸n exponencial, que es com煤n para modelar el tiempo entre eventos en sistemas de servicio.

**Tiempo de servicio** = una distribuci贸n exponencial con un par谩metro **位 (la tasa de servicio)**. 

La distribuci贸n exponencial es adecuada porque describe situaciones en las que los tiempos entre eventos (en este caso, los tiempos de servicio) son independientes y ocurren de manera continua y aleatoria.


#### Proceso:
- Generaci贸n de n煤meros aleatorios: La funci贸n np.random.uniform(0, 1) genera un n煤mero aleatorio uniforme entre 0 y 1.
- Transformaci贸n: Usamos la f贸rmula del m茅todo de la transformada inversa para obtener los tiempos de servicio exponenciales.
- Visualizaci贸n: Se genera un histograma que muestra la distribuci贸n de los tiempos de servicio simulados, lo cual sigue la forma esperada de la distribuci贸n exponencial.

```python
import numpy as np
import matplotlib.pyplot as plt

# Par谩metro de la distribuci贸n exponencial (tasa de servicio)
lambda_servicio = 1 / 2  # Tiempo promedio de servicio = 2 minutos

# Funci贸n para generar un tiempo de servicio
def generar_tiempo_servicio(lambda_servicio):
    u = np.random.uniform(0, 1)  # Generar n煤mero aleatorio uniforme
    tiempo_servicio = -1 / lambda_servicio * np.log(1 - u)  # Aplicar transformada inversa
    return tiempo_servicio

# Generar 100 tiempos de servicio
tiempos_servicio = [generar_tiempo_servicio(lambda_servicio) for _ in range(100)]

# Mostrar los tiempos de servicio
print("Tiempos de servicio generados:", tiempos_servicio)

# Graficar los tiempos de servicio simulados
plt.hist(tiempos_servicio, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribuci贸n de Tiempos de Servicio Simulados (Exponencial)')
plt.xlabel('Tiempo de servicio (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
